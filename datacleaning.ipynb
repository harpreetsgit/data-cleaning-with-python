```python
import pandas as pd
import numpy as np
import random as rnd
```

```python
titanic = pd.read_csv("titanic1.csv")
```

```python
titanic
```

```python
titanic1 = titanic.dropna(axis=1, how ='all’)   # Drop the columns where all elements are missing values
```

```python
titanic2 = titanic.dropna(axis=1, how='any')   # Drop the columns where any of the elements are missing values
```

```python
titanic2
```

```python
titanic3 = titanic2.dropna(axis=0, thresh=2)   # Keep only the rows which contain 2 missing values maximum
```

```python
titanic3
```

```python
titanic4 = titanic3.fillna(titanic3.mean())   # Fill all missing values with the mean of the particular column
```

```python
titanic4
```

```python
titanic5 = titanic4.fillna(titanic4['PassengerId'].median())   # Fill any missing value in column 'A' with the column median
```
